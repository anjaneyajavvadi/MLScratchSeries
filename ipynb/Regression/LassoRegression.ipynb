In lasso regression we try to shrick some weights exactly to zero if w>1 we use +1 if w==0 wwe use 0 else we use -1

```python
class LassoRegression:
    def __init__(self, lr=0.01, n_iters=1000,alpha=0.1):
        self.lr = lr
        self.n_iters = n_iters
        self.alpha=alpha
        self.weights = None
        self.bias = 0

    def fit(self, X, y):
        n_samples,n_features=X.shape
        self.weights=np.zeros(n_features)
        self.bias=0
        for _ in range(self.n_iters):
            y_pred=np.dot(X,self.weights)+self.bias

            dw=(1/n_samples)*np.dot(X.T,(y_pred-y))+self.alpha*np.sign(self.weights)
            db=(1/n_samples)*np.sum(y_pred-y)

            self.weights-=self.lr*dw
            self.bias-=self.lr*db
            print(self.weights)

    def predict(self,X):
        return np.dot(X,self.weights)+self.bias

    def mse(self,y_true,y_pred):
        return np.mean((y_true-y_pred)**2)
```

```python
from sklearn.linear_model import Lasso
from sklearn.datasets import make_regression
from sklearn.model_selection import train_test_split
```

```python
X,y=make_regression(n_samples=300,n_features=5,n_informative=3,noise=20,random_state=42)
```

```python
lasso=LassoRegression(lr=0.01,n_iters=1000,alpha=0.1)
lasso.fit(X,y)
```

```python
lasso.mse(y,lasso.predict(X))
```
